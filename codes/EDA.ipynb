# EDA for Life Expectancy

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
life = pd.read_csv('Life Expectancy Data.csv')
```

```python
life.columns
```

```python
life = life[~life['Life expectancy '].isnull()]
```

## Life expectancy global average life expectancy 2001-2015

```python
plt1 = life[['Year', 'Life expectancy ']]
```

```python
plt1 = plt1.groupby('Year').mean()
```

```python
fig, ax = plt.subplots(1,1, figsize = (10, 6))
ax.plot(plt1)
ax.set_title("Global Average Life Expectancy in 2001-2015")
plt.xlabel('Year', size=10)
plt.ylabel('Life Expectancy', size=10)

plt.show()
```

## Average Life Expectancy between Developed & Developing Countries

```python
plt2 = life[['Status', 'Life expectancy ']]
```

```python
plt2 = plt2.groupby('Status').mean()
```

```python
plt2
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
sns.barplot(list(range(2)), plt2['Life expectancy '])
plt.xticks(ticks=range(2), labels=('Developed', 'Deveploping'))
plt.text(-.05, plt2['Life expectancy ']["Developed"]+.6, int(plt2['Life expectancy ']["Developed"]), size = 12)
plt.text(.95, plt2['Life expectancy ']["Developing"]+.6, int(plt2['Life expectancy ']["Developing"]), size = 12)
ax.set_title('Developed & Developing countries Average Life Expectancy', size = 15)
plt.show()
```
